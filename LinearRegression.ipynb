### **Correlation**

**As we can tell from the analysis ,the values in some columns are more closely related to the values in charges compared to other columns.E.g: age and charges seem to grow together ,whereas bmi and charges don't.**

**This relationship is often expressed numerically using a measure called the correlation coefficient,which can be computed using the '.corr' method of a Pandas series.**

```python
medical_df.charges.corr(medical_df.age)
```

```python
medical_df.charges.corr(medical_df.bmi)
```

**To compute the correlation of categorical columns,they must be converted into numeric columns.**

```python
smoker_values = {'no':0,'yes':1}
smoker_numeric = medical_df.smoker.map(smoker_values)
smoker_numeric
```

```python
medical_df.charges.corr(smoker_numeric)
```

**Pandas dataframe also provide a '.corr' method to compute the correlation coefficient between all pairs of numeric columns.**

```python
medical_df.corr()
```

**The result of '.corr' is called a correlation matrix and is often visualized using a heatmap.**

```python
sns.heatmap(medical_df.corr(),cmap='Reds',annot=True)
plt.title('Correlation Matrix')
```

```python
import numpy as np
```

**Let's use the LinearRegression class from scikit-learn to find the best fit line for age vs charges using the ordinary least square optimization technique.**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# we use the fit method of the model to find the best fit line for the inputs and targest.
help(model.fit)
```

**Some other method**

```python
medical_df['smoker_numeric'] = smoker_numeric
```

```python
medical_df.corr()
```

**We know that the smoker and age columns have the strongest correlaion with charges.Let's create a data frame containing just the data for non-smokers.**

```python
non_smoker_df = medical_df[medical_df.smoker=='no']
```

```python
non_smoker_df
```

```python
# Note that the input must be a 2-d array,so we'll need to pass a dataframe,instead a single columns
inputs = non_smoker_df[['age']]
target = non_smoker_df.charges
print('Inputs Shape:',inputs.shape)
print('Targets Shape:',target.shape)
```

```python
smoker_input = non_smoker_df[['smoker_numeric']]
```

```python
inputs
```

**Fit the model**

```python
model.fit(inputs,target)
```

**We can now make predictions using the model**

```python
model.predict(np.array([[23],[37],[61]]))
```

**Let's compute the prediction for the entire set of inputs**

```python
predictions = model.predict(inputs)
```

```python
predictions
```

**Let's compute the RMSE loss to evaluate the model**

```python
def rmse(targets,predictions):
  return np.sqrt(np.mean(np.square(targets-predictions)))
```

```python
rmse(target,predictions)
```

**The parameters of the model are stored in the coef_ and intercept_ properties**

```python
model.coef_
```

```python
model.intercept_
```

**Visualize the Model**

```python
plt.plot(inputs,predictions,'r',alpha=0.9)
plt.scatter(inputs,target,s=8,alpha=0.8)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend(['Prediction','Actual'])
```
